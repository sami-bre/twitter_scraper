# Check if a result is blog post

````python


gemini_api_key = os.getenv("GEMINI_API_KEY")

def is_content_blog(content):
    url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash:generateContent?key={gemini_api_key}"

    headers = {
        "Content-Type": "application/json"
    }

    

    data = {
        "contents": [
            {
                "parts": [
                    {"text": prompt_template.format(scraped_content=content)}
                ]
            }
        ]
    }

    response = requests.post(url, headers=headers, json=data)
    result = response.json()
    return result
````

````python


def extract_json_data(result: dict) -> dict:
    # Get the text from the first candidate
    json_text = result["candidates"][0]["content"]["parts"][0]["text"]

    # Remove leading markdown formatting (```json) and trailing formatting (```)
    start = json_text.find("```json")
    if start != -1:
        start += len("```json")
    else:
        start = 0
    end = json_text.find("```", start)
    json_str = json_text[start:end].strip()

    # Load and return the JSON data from the extracted string
    return json.loads(json_str)
````

# Check the first result

````python
res = is_content_blog(raw_contents[0]["raw_content"])
print(res)
````

````python
final_result = extract_json_data(res)
pprint(final_result)
````

# Check all results

````python
for idx, item in enumerate(raw_contents):
    
    print(f"Processing raw_content at index {idx}")
    try:
        res = is_content_blog(item["raw_content"])
        final_result = extract_json_data(res)
        final_result["url"] = item["url"]
        pprint(final_result)
        print()
    except Exception as e:
        print(f"Error processing content at index {idx}: {e}")
    
````
